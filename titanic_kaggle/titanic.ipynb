```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree


trainfile = 'train.csv'
testfile = 'test.csv'
dataset = pd.read_csv(trainfile,sep=',',header=0)
testset = pd.read_csv(testfile,sep=',',header=0)

dataset
```

```python
drop_list = ['PassengerId','Name','Ticket','Fare']
```

```python
train = dataset.drop(drop_list, axis=1)
test = testset.drop(drop_list, axis=1)
```

```python
fac_list = ['Sex','Cabin','Embarked']

for f in fac_list:
    train[f] = pd.factorize(train[f])[0]
    test[f] = pd.factorize(test[f])[0]
    
for c in train.columns:
    median_value=train[c].median()
    train[c]=train[c].fillna(median_value)
    
for c in test.columns:
    median_value=test[c].median()
    test[c]=test[c].fillna(median_value)
    
train
```

```python
y_train = train.iloc[:,0].values
x_train = train.iloc[:,1:].values

#y_test = test.iloc[:,0].values #didntgive survive outcome
x_test = test.iloc[:, :].values
```

```python
clf_entropy = DecisionTreeClassifier(criterion="entropy", random_state=100, max_depth=3, min_samples_leaf=5)
clf_entropy.fit(x_train,y_train)
```

```python
y_pred = clf_entropy.predict(x_test)

outcomeset = pd.DataFrame()
outcomeset['PassengerId'] = testset['PassengerId']
outcomeset['Survived'] = y_pred
```

```python
outcomeset.to_csv('outcome.csv', index=False)
```
